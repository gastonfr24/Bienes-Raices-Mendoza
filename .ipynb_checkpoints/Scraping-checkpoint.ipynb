<h1 style="color: green;">Función para scrapear</h1>
Ahora que sabemos que trae todo de forma correcta hacemos la función para traer toda la data. <br>
Puedes cambiar el url y el limite de paginación 


```python
# Mendoza
url_path = 'https://www.inmoclick.com.ar/inmuebles/casas-en-venta-en-capital-de-mendoza?favoritos=0&limit=24&prevEstadoMap=&lastZoom=13&precio%5Bmin%5D=&precio%5Bmax%5D=&moneda=1&sup_cubierta%5Bmin%5D=&sup_cubierta%5Bmax%5D=&sup_total%5Bmin%5D=&sup_total%5Bmax%5D=&tiene_precio=on&page='

# Godoy Cruz
#url_path = 'https://www.inmoclick.com.ar/inmuebles/casas-en-venta-en-godoy-cruz-de-mendoza?favoritos=0&limit=24&prevEstadoMap=&lastZoom=13&precio%5Bmin%5D=&precio%5Bmax%5D=&moneda=1&sup_cubierta%5Bmin%5D=&sup_cubierta%5Bmax%5D=&sup_total%5Bmin%5D=&sup_total%5Bmax%5D=&tiene_precio=on&page='

# Mendoza
pagination_limit = 31

# Godoy Cruz
#pagination_limit = 32

all_data = {}

price_USD = []
mts_squared_str = []
house_squared_str = []
all_bathrooms_str = []
all_bedrooms_str = []
address_str = []
country_str = []

house={}
null_data = 0

articles = '//div[@class="row"]/div[@class="col-xs-12 col-sm-6 col-md-3 col-lg-2"]/article'
def get_data(url, limit):
    null_data = 0
    for page in range(limit):
        print(url + str(page))
        
        r = requests.get(url + str(page))

        if r.status_code == 200:
            
            first_page = r.content.decode('utf-8')
            parser = html.fromstring(first_page)
            
            len_articles = parser.xpath(articles)
            
            for index in range(len(len_articles)):
                article = f'//div[@class="row"]/div[@class="col-xs-12 col-sm-6 col-md-3 col-lg-2"]/article[@index="{index}"]//span/text()'
                article_info = parser.xpath(article)
                
                price_article = f'//div[@class="row"]/div[@class="col-xs-12 col-sm-6 col-md-3 col-lg-2"]/article[@index="{index}"]//p[@class="price"]/text()'
                price_info = parser.xpath(price_article)
                if len(article_info) == 6:
                    price_USD.append(price_info[0])
                    country_str.append(article_info[0])
                    address_str.append(article_info[1])
                    mts_squared_str.append(article_info[2])
                    house_squared_str.append(article_info[3])
                    all_bedrooms_str.append(article_info[4])
                    all_bathrooms_str.append(article_info[5])
                else:
                    null_data += 1
    print(null_data)          
    
    
    house['Country'] = country_str
    house['Address'] = address_str
    house['mts2'] = mts_squared_str
    house['mts2_house'] = house_squared_str
    house['Bathrooms'] = all_bathrooms_str
    house['Bedrooms'] = all_bedrooms_str
    house['Price'] = price_USD
    return house
```

```python
house_df = get_data(url_path, pagination_limit)
```

```python
house_data = pd.DataFrame(house_df)
```
